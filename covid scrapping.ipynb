```python
from bs4 import BeautifulSoup
import requests

url=BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
```

```python
url
```

```python
response=requests.get(url)
data=response.text
data=BeautifulSoup(data,'lxml')
data
```

```python
tag=data.table
tag
```

```python
clear=tag.find_all('tr')
clear=clear[9:-8]
clear
```

```python
final=[]
for rows in clear:
    cells=rows.find_all('td')[15:]
    final.append(cells)
final
```

```python
final_2=list(set(clear)-set(final))

final_2
    
```

```python
#splitting from n
final_data=[]
for i in final:
    splitted= i.split('\n')[1:-1]
    final_data.append(splitted)
final_data
```
