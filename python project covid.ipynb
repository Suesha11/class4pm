```python
from bs4 import BeautifulSoup
import requests

url=BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
```

```python
response=requests.get(url)
data=response.text
data=BeautifulSoup(data,'lxml')
data
```

```python
tag=data.table
clear=tag.find_all('tr')
```

```python
clear[9:-8]
```

```python
final=[]
for i in clear:
    final.append(i.get_text())
final
```

```python
final_list=[]
for i in final:
    final_list.append(i.split('\n')[1:-1])
final_list
```

```python
del final_list[1:9]
final_list
```

```python
final_option=[]
for i in final_list:
    final_option.append(i[0:-2])
final_option
    
```

```python
del final_option[0][14:16]
final_option
```

```python
import csv

fields=final_option

with open('covid.csv','a') as f:
    x=csv.writer(f)
    
    for i in fields:
        x.writerow(i)
```
